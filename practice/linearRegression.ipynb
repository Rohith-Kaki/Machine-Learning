```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('R:\Machine Learning\data\canada_per_capita_income.csv')
df.rename(columns={'year':'year','per capita income (US$)': 'pci'},inplace=True)
df.head()
```

```python
%matplotlib inline
plt.xlabel("year")
plt.ylabel("pci")
plt.scatter(df.year,df.pci)
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']],df.pci)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[2020]])
```
